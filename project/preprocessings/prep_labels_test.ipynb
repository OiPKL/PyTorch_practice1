```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def set_file_path(input_file):
    if 11 <= input_file <= 17:
        prep_data_dir = r"../dataset/prep_data/lettuce/labels/training"
    else:
        prep_data_dir = r"../dataset/prep_data/lettuce/labels/validation"

    if input_file == 11:
        vegetableTotalDays = 37
    elif input_file == 12:
        vegetableTotalDays = 58
    elif input_file == 13:
        vegetableTotalDays = 58
    elif input_file == 14:
        vegetableTotalDays = 54
    elif input_file == 15:
        vegetableTotalDays = 56
    elif input_file == 16:
        vegetableTotalDays = 58
    elif input_file == 17:
        vegetableTotalDays = 56
    elif input_file == 18:
        vegetableTotalDays = 58
    elif input_file == 19:
        vegetableTotalDays = 58
    elif input_file == 20:
        vegetableTotalDays = 56
    else:
        vegetableTotalDays = 0

    file_path = os.path.join(prep_data_dir, f"{input_file}_{vegetableTotalDays}.csv")
    return file_path

def making_plot(file_path):
    file_path = set_file_path(file_path)

    data = pd.read_csv(file_path)

    plt.figure(figsize=(50, 12))
    colors = ['red', 'blue', 'green', 'purple', 'orange']
    linewidths = [3, 3, 3, 3, 5]
    normalized_data = (data - data.min()) / (data.max() - data.min())
    for col, color, linewidth in zip(['gardenTemp', 'gardenHumid', 'gardenLight', 'gardenCO2', 'vegetableGrowth'], colors, linewidths):
        plt.plot(normalized_data['index'][::20], normalized_data[col][::20], label=col, color=color, linewidth=linewidth)
    plt.show()

    plt.figure(figsize=(25, 8))
    plt.plot(data['index'][::5], data['gardenTemp'][::5], label='gardenTemp', color='red', linewidth=3)
    plt.tick_params(axis='x', labelsize=10)
    plt.tick_params(axis='y', labelsize=10)
    plt.legend(fontsize=20)
    plt.show()

    plt.figure(figsize=(25, 8))
    plt.plot(data['index'][::5], data['gardenHumid'][::5], label='gardenHumid', color='blue', linewidth=3)
    plt.tick_params(axis='x', labelsize=10)
    plt.tick_params(axis='y', labelsize=10)
    plt.legend(fontsize=20)
    plt.show()

    plt.figure(figsize=(25, 8))
    plt.plot(data['index'][::5], data['gardenLight'][::5], label='gardenLight', color='green', linewidth=3)
    plt.tick_params(axis='x', labelsize=10)
    plt.tick_params(axis='y', labelsize=10)
    plt.legend(fontsize=20)
    plt.show()

    plt.figure(figsize=(25, 8))
    plt.plot(data['index'][::5], data['gardenCO2'][::5], label='gardenCO2', color='purple', linewidth=3)
    plt.tick_params(axis='x', labelsize=10)
    plt.tick_params(axis='y', labelsize=10)
    plt.legend(fontsize=20)
    plt.show()

    plt.figure(figsize=(25, 8))
    plt.plot(data['index'][::5], data['vegetableGrowth'][::5], label='vegetableGrowth', color='orange', linewidth=3)
    plt.tick_params(axis='x', labelsize=10)
    plt.tick_params(axis='y', labelsize=10)
    plt.legend(fontsize=20)
    plt.show()

    print(file_path)
```
